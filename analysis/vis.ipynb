```python
sign_map = np.fromfile("../build/test/sign_map.uint8",dtype=np.int8).reshape(256,384,384)
sign_orig = np.fromfile("../build/sign.int8",dtype=np.int32).reshape(256,384,384)
slice_select = 35
x1 = 190
x2 = 220
y1 = 0
y2 = 5
figs,axs = plt.subplots(1,4) 

axs[0].imshow(sign_orig[slice_select,x1:x2, y1:y2],cmap ="coolwarm",vmin=-1,vmax=1)
axs[0].set_title("Original")
axs[1].imshow(sign_map[slice_select,x1:x2, y1:y2],cmap ="coolwarm",vmin=-1,vmax=1)
axs[1].set_title("cuda")
axs[2].imshow(sign_map[slice_select,x1:x2, y1:y2]-sign_orig[slice_select,x1:x2, y1:y2],cmap ="coolwarm",vmin=-1,vmax=1)
axs[3].imshow(orig_bound[slice_select,x1:x2, y1:y2],cmap ="coolwarm")
print(np.sum(sign_map!=sign_orig))
# print(sign_map[slice_select,x1:x2, y1:y2]-sign_orig[slice_select,x1:x2, y1:y2])
```
